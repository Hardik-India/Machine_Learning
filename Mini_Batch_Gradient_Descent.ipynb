```python
from sklearn.datasets import load_diabetes

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import r2_score
```

```python
x,y = load_diabetes(return_X_y=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
reg = LinearRegression()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
print("Betas value : ",reg.coef_)
print("Intercept value : ",reg.intercept_)
```

```python
import random
class MBGDRegressor:

  def __init__(self,batch_size,learning_rate,epochs):
    self.coef_ = None
    self.intercept_ = None
    self.lr = learning_rate
    self.epochs = epochs
    self.batch_size = batch_size

  def fit(self,x_train,y_train):
    # initialize the coef with all ones and the intercept with 0
    self.intercept_ = 0
    self.coef_ = np.ones(x_train.shape[1])

    for i in range(self.epochs):
      for j in range(int(x_train.shape[0]/self.batch_size)):

        # Generating Random numbers based on batch size
        idx = random.sample(range(x_train.shape[0]),self.batch_size)

        y_hat = np.dot(x_train[idx],self.coef_) + self.intercept_

        # calculating the intercept values
        intercept_der = -2 * np.mean(y_train[idx] - y_hat)
        self.intercept_ = self.intercept_ - (self.lr * intercept_der)

        # calculating the coefficient values
        coef_der = -2 * np.dot((y_train[idx] - y_hat),x_train[idx])
        self.coef_ = self.coef_ - (self.lr * coef_der)

  def predict(self,x_test):
    return np.dot(x_test,self.coef_) + self.intercept_
```

```python
mbr = MBGDRegressor(batch_size=int(x_train.shape[0]/10),learning_rate=0.01,epochs=500)
```

```python
mbr.fit(x_train,y_train)
```

```python
mbr.coef_
```

```python
mbr.intercept_
```

```python
y_pred = mbr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
