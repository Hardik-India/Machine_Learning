```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Let's apply OLS
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, lr.predict(x),color='red')
```

```python
# Let's apply gradient descent using m=78.5 and b=0(any random value)
y_pred1 = ((78.35 * x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred,color='red', label='OLS')
plt.plot(x, y_pred1, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration-1
m=78.35
b=0

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
# Let's take learning rate = 0.1
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
```

```python
b
```

```python
y_pred2 = ((78.35 * x)+b).reshape(4)
```

```python
y_pred2
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred,color='red', label='OLS')
plt.plot(x, y_pred2, color='blue', label='b={}'.format(b))
plt.plot(x, y_pred1, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration-2

loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3 = ((78.35 * x)+b).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred,color='red', label='OLS')
plt.plot(x, y_pred2, color='blue', label='b={}'.format(b))
plt.plot(x, y_pred1, color='green', label='b=0')
plt.plot(x, y_pred3, color='black', label='b={}'.format(b))
plt.legend()
plt.show()
```

## **Performing Gradient Descent using Loop**

```python
b=0
m=78.35
lr=0.03
epochs=100

for i in range(epochs):
  loss_slope = -2*np.sum(y-m*x.ravel()-b)
  b=b-(lr*loss_slope)
  print(b)
  y_pred = m*x+b
  plt.plot(x,y_pred)
plt.scatter(x,y)
```
